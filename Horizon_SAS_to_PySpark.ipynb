```python
# we have to confirm date type in table"TDPRD.SMC_FACT_PHRM_CLM" in column "PRSC_DT"

"""
proc sql;
Create table TESTa AS SELECT DIM_MBR_SK, CUST_CARD_ID, PAT_FRST_NM, PAT_LST_NM, PAT_BRTH_DT, PHRM_SK, CLM_SK_RX, CLM_RX_ID, ORG_SRC_SYS_CD, PHRM_CLM_STAT_CD, INCR_DT,
AGE,

 PAT_SEX_CD, PDT_NM_GEN, PDT_NM_RX, DEA_CD_DSC, PRSB_ADDR_LN, PHRM_NPI_ID, PRSB_NPI_ID, PRSB_ST_NM, MKT_SEG_NM, MAIN_GRP_ID, SGRP_ID, GRP_PKG_PDT_CD
FROM
(SELECT distinct DIM_RX_PDT_SK,
PDT_NM_GEN,
PDT_NM_RX,
DEA_CD_DSC
FROM TDPRD.SMC_DIM_PDT_RX

 WHERE
PDT_GEN_ID in (&OPI) or PDT_GEN_ID in (&MAT) or PDT_GEN_ID in (&STIM)
) DIM
INNER JOIN
(select distinct PDT_RX_SK, DIM_MBR_SK, CUST_CARD_ID, PAT_FRST_NM, PAT_LST_NM, PAT_BRTH_DT, PHRM_NPI_ID, PRSB_ADDR_LN, CLM_SK_RX,
CLM_RX_ID, PHRM_SK, PAT_SEX_CD, PRSB_NPI_ID,

 INCR_DT,
PRSB_ST_NM,
floor(((today() - PAT_BRTH_DT)/365.25)) as AGE, PHRM_CLM_STAT_CD,
ORG_SRC_SYS_CD,
MKT_SEG_NM,
MAIN_GRP_ID, SGRP_ID, GRP_PKG_PDT_CD
FROM TDPRD.SMC_FACT_PHRM_CLM WHERE
PHRM_CLM_STAT_CD IN ('P', 'R') AND PRSC_DT BETWEEN
&BEG12 /*&BEG1 */ AND &END1
AND ORG_SRC_SYS_CD = 'CLMPRM'
and CURR_RCD_IND in ('Y') /*****/
and MKT_SEG_NM not in ('SRMRKT','SRMRKT-MA ONLY')
) CLM

 ON dim.DIM_RX_PDT_SK = clm.PDT_RX_SK
WHERE DIM_MBR_SK ne -1; quit;
"""


OPI = [1, 2, 3] # OPI values
MAT = [4, 5, 6] # MAT values
STIM = [7, 8, 9] # STIM values


dim = (
    spark.table("TDPRD.SMC_DIM_PDT_RX") \
         .filter(col("PDT_GEN_ID").isin(OPI) | col("PDT_GEN_ID").isin(MAT) | col("PDT_GEN_ID").isin(STIM)) \
         .select("DIM_RX_PDT_SK", "PDT_NM_GEN", "PDT_NM_RX", "DEA_CD_DSC") \
         .distinct()
      )

clm = spark.table("TDPRD.SMC_FACT_PHRM_CLM") \
           .filter(col("PHRM_CLM_STAT_CD").isin(["P", "R"])) \
           .filter(col("PRSC_DT").between(BEG12, END1)) \ 
           .filter(col("ORG_SRC_SYS_CD") == "CLMPRM") \
           .filter(col("CURR_RCD_IND") == "Y") \
           .filter(~col("MKT_SEG_NM").isin(["SRMRKT", "SRMRKT-MA ONLY"])) \
           .select("PDT_RX_SK", "DIM_MBR_SK", "CUST_CARD_ID", "PAT_FRST_NM", 
                   "PAT_LST_NM", "PAT_BRTH_DT", "PHRM_NPI_ID", "PRSB_ADDR_LN", "CLM_SK_RX", "CLM_RX_ID", "PHRM_SK",
                   "PAT_SEX_CD", "PRSB_NPI_ID", "INCR_DT", "PRSB_ST_NM",
                   floor(((current_date() - col("PAT_BRTH_DT"))/365.25)).alias("AGE"), "PHRM_CLM_STAT_CD", "ORG_SRC_SYS_CD", 
                   "MKT_SEG_NM", "MAIN_GRP_ID", "SGRP_ID", "GRP_PKG_PDT_CD") \
           .distinct()
    
TESTa = dim.join(clm, col("DIM_RX_PDT_SK") == col("PDT_RX_SK"), "inner") \
           .filter(col("DIM_MBR_SK") != -1) \
           .select("DIM_MBR_SK", "CUST_CARD_ID", "PAT_FRST_NM", "PAT_LST_NM", "PAT_BRTH_DT", "PHRM_SK", "CLM_SK_RX",
                   "CLM_RX_ID", "ORG_SRC_SYS_CD", "PHRM_CLM_STAT_CD", "INCR_DT", "AGE", "PAT_SEX_CD", "PDT_NM_GEN", 
                   "PDT_NM_RX", "DEA_CD_DSC", "PRSB_ADDR_LN", "PHRM_NPI_ID", "PRSB_NPI_ID", "PRSB_ST_NM", "MKT_SEG_NM", 
                   "MAIN_GRP_ID", "SGRP_ID", "GRP_PKG_PDT_CD")
```

```python
# gather 6 months data

dim1 = spark.table("TDPRD.SMC_DIM_PDT_RX").select("DIM_RX_PDT_SK", "PDT_NM_GEN", "PDT_NM_RX", "DEA_CD_DSC") \
            .filter(col("PDT_GEN_ID").isin(OPI) | col("PDT_GEN_ID").isin(MAT) | col("PDT_GEN_ID").isin(STIM)) \
            .distinct()

clm1 = spark.table("TDPRD.SMC_FACT_PHRM_CLM") \
            .filter(col("PHRM_CLM_STAT_CD").isin(["P", "R"])) \
            .filter(col("PRSC_DT").between(BEG6, END1)) \
            .filter(col("ORG_SRC_SYS_CD") == "CLMPRM") \
            .filter(col("CURR_RCD_IND") == "Y") \
            .filter(~col("MKT_SEG_NM").isin(["SRMRKT", "SRMRKT-MA ONLY"])) \
            .select("PDT_RX_SK", "DIM_MBR_SK", "CUST_CARD_ID", "PAT_FRST_NM", 
                    "PAT_LST_NM", "PHRM_NPI_ID", "PRSB_ADDR_LN", "CLM_SK_RX", "CLM_RX_ID", "PHRM_SK",
                    "PAT_SEX_CD", "PRSB_NPI_ID", "INCR_DT", "PRSB_ST_NM",
                    floor(((current_date() - col("PAT_BRTH_DT"))/365.25)).alias("AGE"),"PHRM_CLM_STAT_CD", "ORG_SRC_SYS_CD", 
                    "MKT_SEG_NM", "MAIN_GRP_ID", "SGRP_ID", "GRP_PKG_PDT_CD") \
            .distinct()


SIXMO = dim1.join(clm1, col("DIM_RX_PDT_SK") == col("PDT_RX_SK"), "inner") \
            .filter(col("DIM_MBR_SK") != -1) \
            .select("DIM_MBR_SK", "CUST_CARD_ID", "PAT_FRST_NM", "PAT_LST_NM", "PHRM_SK", "CLM_SK_RX",
                    "CLM_RX_ID", "ORG_SRC_SYS_CD", "PHRM_CLM_STAT_CD", "INCR_DT", "AGE",
                    "PAT_SEX_CD", "PDT_NM_GEN", "PDT_NM_RX", "DEA_CD_DSC", "PRSB_ADDR_LN",
                    "PHRM_NPI_ID", "PRSB_NPI_ID", "PRSB_ST_NM", "MKT_SEG_NM", "MAIN_GRP_ID",
                    "SGRP_ID", "GRP_PKG_PDT_CD") 
```

```python
"""
data SIXMOa (keep=DIM_MBR_SK);
set SIXMO;
run;
proc sort data=SIXMOa nodupkey;
by DIM_MBR_SK;
run;
"""

# SIXMOa = SIXMO.select("DIM_MBR_SK").distinct().sort(col("DIM_MBR_SK")).rdd.map(lambda r: r[0]).collect()
SIXMOa = SIXMO.select("DIM_MBR_SK").distinct().sort(col("DIM_MBR_SK"))
```

```python
"""
proc sort data=testa;
by DIM_MBR_SK;
run;
"""

TESTa = TESTa.orderBy("DIM_MBR_SK")
```

```python
"""
data testb;
merge testa (in=a) SIXMOa (in=b);
by DIM_MBR_SK;
if a = b;
run;
"""
# sixmo_unique_dim_mbr = .select('DIM_MBR_SK').distinct().rdd.map(lambda r: r[0]).collect()
testb = TESTa.join(SIXMOa, on="DIM_MBR_SK", how="inner")
# testb = TESTa.filter(col("DIM_MBR_SK").isin(SIXMOa))
```

```python
"""
data testb1 (keep=DIM_MBR_SK PRSB_NPI_ID);
set testb;
run;
proc sort data=testb1 nodupkey;
by DIM_MBR_SK PRSB_NPI_ID;
run;
"""

testb1 = testb.select("DIM_MBR_SK", "PRSB_NPI_ID").distinct().orderBy("DIM_MBR_SK", "PRSB_NPI_ID")
```

```python
"""                                                                     
proc sql;
create table testb1_a
as select
DIM_MBR_SK,
PRSB_NPI_ID,
count(DIM_MBR_SK) as countmem
from testb1
group by DIM_MBR_SK;
quit;
"""


testb1_a = testb1.groupby("DIM_MBR_SK") \                        
                 .agg(count("PRSB_NPI_ID").alias("countmem"))  
```

```python
"""
data testb1_b (keep=DIM_MBR_SK);
set testb1_a;
if countmem = 1;
run;
proc sort data=testb1_b;
by DIM_MBR_SK;
run;
"""

testb1_b = testb1_a.filter(col("countmem") == 1).select("DIM_MBR_SK").sort("DIM_MBR_SK")
```

```python
"""
proc sort data=testb;
by DIM_MBR_SK;
run;
"""

testb = testb.orderBy("DIM_MBR_SK")
```

```python
"""
data testb_clean;
merge testb (in=a) testb1_b (in=b);
by DIM_MBR_SK;
if a and not b;
run;
"""

testb_clean = testb.join(testb1_b,  on="DIM_MBR_SK",  how="left_anti")
```

```python
"""
data testb2 (keep=DIM_MBR_SK PRSB_ADDR_LN);
set testb;
run;
proc sort data=testb2 nodupkey;
by DIM_MBR_SK PRSB_ADDR_LN;
run;
"""

testb2 = testb.select("DIM_MBR_SK", "PRSB_ADDR_LN").distinct().orderBy("DIM_MBR_SK", "PRSB_ADDR_LN")
```

```python
"""
proc sql;
create table testb2_a
as select distinct
DIM_MBR_SK,
PRSB_ADDR_LN,
count(DIM_MBR_SK) as countmem
from testb2
group by DIM_MBR_SK;
quit;
"""


testb2_a = testb2.groupby("DIM_MBR_SK") \                        
                 .agg(count("PRSB_ADDR_LN").alias("countmem"))
```

```python
"""
data testb2_b (keep=DIM_MBR_SK);
set testb2_a;
if countmem = 1;
run;
proc sort data=testb2_b nodupkey;
by DIM_MBR_SK;
run;
"""

testb2_b = testb2_a.filter(col("countmem") == 1).select("DIM_MBR_SK").sort("DIM_MBR_SK").distinct()
```

```python
"""
proc sort data=testb_clean;
by DIM_MBR_SK;
run;
"""

testb_clean = testb_clean.orderBy("DIM_MBR_SK")
```

```python
"""
data testb_cleaner;
merge testb_clean (in=a) testb2_b (in=b);
by DIM_MBR_SK;
if a and not b;
run;
"""

testb_cleaner = testb_clean.join(testb2_b,  on="DIM_MBR_SK",  how="left_anti")
```

```python
"""
data testb_cleaner_a;
set testb_cleaner;
md_npi = input(prsb_npi_id,12.);
run;
"""

testb_cleaner_a = testb_cleaner.withColumn("md_npi", col("PRSB_NPI_ID").cast("double"))
```

```python
"""
proc sql;
create table test
as select
b.practice_name,
a.*

from testb_cleaner_a a left join AMY.MDPRACTICELIST b
on a.md_npi = b.md_npi;
quit;
"""
amy_practice_table=spark.table("AMY.MDPRACTICELIST")

test = testb_cleaner_a.join(amy_practice_table, on="md_npi", how="left").select("practice_name", *testb_cleaner_a.columns)
```

```python
"""
data onlist notonlist;
set test;
if practice_name ne '' then output onlist;
else output notonlist;
run;
"""

onlist = test.filter(col("practice_name") != "")
notonlist = test.filter(col("practice_name") == "")
```

```python
"""
proc sort data=onlist nodupkey;
by dim_mbr_sk;
run;
"""

onlist = onlist.orderBy("DIM_MBR_SK").dropDuplicates("DIM_MBR_SK")
```

```python
"""
data all_interim;
set onlist notonlist;
run;
"""

all_interim = onlist.union(notonlist)
```

```python
"""
proc sort data=all_interim;
by dim_mbr_sk;
run;
"""

all_interim = all_interim.orderBy(col("DIM_MBR_SK"))
```

```python
"""
proc sql;
create table all_interim_a
as select
dim_mbr_sk,
count(dim_mbr_sk) as countmem
from all_interim
group by dim_mbr_sk;
quit;
"""

all_interim_a = all_interim.groupby("DIM_MBR_SK")\
                           .agg(count("DIM_MBR_SK").alias("countmem"))
```

```python
"""
data all_interim_b (keep=dim_mbr_sk);
set all_interim_a;
if countmem = 1;
run;
proc sort data=all_interim_b ;
by dim_mbr_sk;
run;
"""

all_interim_b = all_interim_a.filter(col("countmem") == 1).select("DIM_MBR_SK").orderBy("DIM_MBR_SK")
```

```python
"""
proc sort data=all_interim;
by dim_mbr_sk;
run;
"""
# It is repeated and has no meaning. Check two code block above.
```

```python
"""
data testb_cleanest;
merge all_interim (in=a) all_interim_b (in=b);
by dim_mbr_sk;
if a and not b;
run;
"""

testb_cleanest = all_interim.join(all_interim_b,  on="DIM_MBR_SK",  how="left_anti")
```

```python
"""
data practice_count;
set testb_cleanest;
indpractice = '';
if practice_name ne '' then indpractice = practice_name;
else indpractice = PRSB_NPI_ID||PRSB_ADDR_LN; /*Unique combinations*/
run;
"""

practice_count = testb_cleanest.withColumn("indpractice", when(col("practice_name") != "", col("practice_name")).otherwise(concat(col("PRSB_NPI_ID"), col("PRSB_ADDR_LN"))))
```

```python
"""
proc sort data=practice_count  nodupkey;
by dim_mbr_sk indpractice;
run;
"""

practice_count = practice_count.orderBy("DIM_MBR_SK", "indpractice").dropDuplicates(["dim_mbr_sk", "indpractice"])
```

```python
"""
proc sql;
create table sum
as select
dim_mbr_sk,
count(indpractice) as count_of_practices
from practice_count
group by dim_mbr_sk;
quit;
"""

sum = practice_count.groupby("DIM_MBR_SK") \
                    .agg(count("indpractice").alias("count_of_practices")) \
```

```python
"""
proc sql;
create table amy.testb_cleanest1
as select
a.*,
b.count_of_practices
from testb_cleanest as a
left join sum as b
on a.dim_mbr_sk = b.dim_mbr_sk;
quit;
"""

testb_cleanest1 = testb_cleanest.join(sum, on="DIM_MBR_SK", how="left").select(*testb_cleanest.columns, "count_of_practices")  # It is present in amy library in sas
```
